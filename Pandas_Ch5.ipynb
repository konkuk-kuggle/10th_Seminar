# 05-1. 분석하기 좋은 데이터



* 깔끔한 데이터 (Tidy Data)
  * 분석 목적에 맞는 데이터를 모아 새로운 표 형성
  * 측정한 값은 행(row)을 구성
  * 변수는 열(column)로 구성

EX) 첨단 산업 기업의 주식 가격에 데이터를 분석   
→ '기업 정보'데이터 집합에서 첨단 기술 기업 추출   
→ '주식 가격' 데이터 집합에서 해당 기업들의 측정 값 추출   
  ⇒ 두 데이터 집합 연결

##### ② 좌우 연결 - 열방향

```python
col_concat = pd.concat([df1, df2, df3], axis=1)
print(col_concat)
```

```python
print(col_concat['A'])
```

```python
# 시리즈는 column 방향이므로, 앞과 달리 데이터 프레임(transpose)으로 변경할 필요 X
col_concat['new_col_list'] = ['n1', 'n2', 'n3', 'n4']
print(col_concat)
```

```python
print(pd.concat([df1, df2, df3], axis=1, ignore_index=True))
```

##### ③ 공통 열/인덱스만 연결

```python
# 열이름 설정
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']
print(df1)
print(type(df1))
```

```python
print(df2)
print(type(df2))
```

```python
print(df3)
print(type(df3))
```

```python
row_concat = pd.concat([df1, df2, df3])
print(row_concat)
```

```python
# 공통 열만 연결 - 모두 만족하는 열은 없음.
print(pd.concat([df1, df2, df3], join='inner'))
```

```python
# df1, df3의 공통 열 연결
print(pd.concat([df1, df3], ignore_index=False, join='inner'))
```

```python
# 행 이름 설정 (index)
df1.index = [0, 1, 2, 3]
df2.index = [4, 5, 6, 7]
df3.index = [0, 2, 5, 7]

print(df1, '\n')
print(df2, '\n')
print(df3, '\n')
```

```python
col_concat = pd.concat([df1, df2, df3], axis=1)
print(col_concat)
```

```python
print(pd.concat([df1, df3], axis=1, join='inner'))
```

```python
print(pd.concat([df1, df3], axis=1, join='outer')) #default
```

# 05-3. 데이터 연결 마무리

```python
person = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kuggle/data/data/survey_person.csv')
site = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kuggle/data/data/survey_site.csv')
survey = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kuggle/data/data/survey_survey.csv')
visited = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kuggle/data/data/survey_visited.csv')
```

```python
print(person, '\n')
print(site, '\n')
print(survey, '\n')
print(visited, '\n')
```

```python
visited_subset = visited.loc[[0, 2, 6], ]
```

```python
# 기본 - 내부 조인
# site(left)의 name column = visited_subset(right)의 site column
o2o_merge = site.merge(visited_subset, left_on='name', right_on='site')
print(o2o_merge)
```

```python
m2o_merge = site.merge(visited, left_on='name', right_on='site')
print(m2o_merge)
# site의 row가 반복됨. (DR-1, DR-3)
```

```python
ps = person.merge(survey, left_on='ident', right_on='person')
vs = visited.merge(survey, left_on='ident', right_on='taken')

print(ps, '\n')
print(vs)
```

```python
print(ps.columns)
print(vs.columns)
```

```python
ps_vs = ps.merge(vs,
                 left_on=['ident', 'taken', 'quant', 'reading'],
                 right_on=['person', 'ident', 'quant', 'reading'])
# 기준이 여러개
# 이름이 같을 시 - a_x : left df / b_y : right df
# 중복되는 기준은 병합

print(ps_vs.loc[0, ]) # 첫번째 행만 확인
```
