### 1) 누락값과 누락값 확인하기

```python
from numpy import NaN, NAN, nan
import pandas as pd
```

```python
print(NaN == True)
print(NaN == False)
print(NaN == 0)
print(NaN == '')
# 0, '', True, False 등 모두 아님. - missing value
```

```python
print(NaN==NaN)
print(NaN==nan)
print(NaN==NAN)
print(nan==NAN)
```

```python
print(pd.isnull(NaN))
print(pd.isnull(nan))
print(pd.isnull(NAN))
```

```python
print(pd.notnull(NaN))
print(pd.notnull(nan))
print(pd.notnull(NAN))
```

```python
life_exp = gapminder.groupby(['year'])['lifeExp'].mean()
print(life_exp) #1952~2007
```

```python
'''
print(life_exp.loc[range(2000, 2010), ])
'''
# row만 지정 - 모든 col 표시
# 오류 발생
# 빈 DataFrame 생성


import pandas as pd
import numpy as np

# life_exp이 Series라고 가정
df = pd.DataFrame()

for i in range(2000, 2010):
    try:
        df = pd.concat([df, pd.DataFrame([life_exp.loc[i]], index=[i])])
    except KeyError:
        df = pd.concat([df, pd.DataFrame([np.nan], index=[i])])

print(df)
```

```python
import pandas as pd
import numpy as np

# life_exp이 Series라고 가정
df = pd.DataFrame(columns=['year', 'value'])

for i in range(2000, 2010):
    try:
        # Series를 DataFrame으로 변환하고 'year' 열을 추가
        row = pd.DataFrame({'year': [i], 'value': [life_exp.loc[i]]})
    except KeyError:
        # 해당 연도가 없을 때 NaN으로 처리
        row = pd.DataFrame({'year': [i], 'value': [np.nan]})

    df = pd.concat([df, row], ignore_index=True)

print(df)
```

### 4) 누락값 처리하기

```python
print(ebola.fillna(0).head().iloc[:,0:5]) # 누락값을 0으로 변경
```

```python
print(ebola.fillna(method='ffill').head().iloc[:,0:5])
# 누락값이 나타나기 전의 행의 값으로 누락값 변경
```

```python
print(ebola.fillna(method='bfill').head().iloc[:,0:5])
# 누락값이 나타난 후의 행의 값으로 누락값 변경
```

```python
print(ebola.interpolate().head().iloc[:,0:5])
# 누락값 양쪽에 있는 값을 이용하여 중간값으로 누락값 변경
```

```python
print(ebola.shape)
print()

ebola_dropna = ebola.dropna()
print(ebola_dropna.shape)
```

##### 누락값이 포함된 데이터 계산하기

```python
ebola['Cases_multiple'] = ebola['Cases_Guinea'] + ebola['Cases_Liberia'] + ebola['Cases_SierraLeone']
# 새로운 column 형성
# 셋 중에 누락값이 하나라도 있으면, NaN으로 도출

ebola_subset = ebola.loc[:, ['Cases_Guinea', 'Cases_Liberia', 'Cases_SierraLeone', 'Cases_multiple']]
print(ebola_subset.head(n=10))
```

```python
print(ebola.Cases_Guinea.sum(skipna = True)) # 누락값 무시하여 column total value 계산
print()
print(ebola.Cases_Guinea.sum(skipna = False)) # 누락값 포함하여 계산 - NaN
```
