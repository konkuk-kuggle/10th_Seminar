##ì¡°ì‚¬, ì–´ë¯¸, êµ¬ë‘ì  ì‚­ì œí•˜ëŠ” í•¨ìˆ˜ ë§Œë“¤ì–´ë³´ê¸°
**KoNLPy í•œêµ­ì–´ í’ˆì‚¬ íƒœê¹…í‘œ:**https://docs.google.com/spreadsheets/d/1OGAjUvalBuX-oZvZ_-9tEfYD2gQe7hTGsgUpiiBSXI8/edit?gid=0#gid=0

* ìœ„ ì£¼ì†Œì˜ íƒœê¹…í‘œë¥¼ ì°¸ê³ í•˜ì—¬ ì‹¤ìŠµí•´ë³´ì‹œê¸° ë°”ëë‹ˆë‹¤.
* ì•„ë˜ ì‘ì„±ëœ ì½”ë“œëŠ” í˜•íƒœì†Œ ë¶„ì„ê¸°ë¥¼ í†µí•´ ì§€ì • ëœ í’ˆì‚¬ë¥¼ ì‚­ì œí•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“  ê²ƒ ì…ë‹ˆë‹¤.
* KoNLPy í’ˆì‚¬ íƒœê¹…í‘œë¥¼ í™•ì¸í•˜ì—¬ ì—¬ëŸ¬ í’ˆì‚¬ë¥¼ ì‚­ì œí•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
***oktëŠ” í’ˆì‚¬ íƒœê¹…í‘œì˜ Twitter Korean Textë¥¼ ì°¸ê³ í•˜ì‹œë©´ ë©ë‹ˆë‹¤.**

```python
# í˜•íƒœì†Œ ë¶„ì„ê¸°(Okt) ë¶ˆëŸ¬ì˜¤ê¸°

def okt_clean(text):
    clean_text = []
    okt_pos = okt.pos(text, stem=True)
    for txt, pos in okt_pos:
        if pos not in ['â­ï¸','â­ï¸','â­ï¸']: # KoNLPy í’ˆì‚¬ íƒœê¹…í‘œ ì°¸ê³ í•˜ì—¬ ì¡°ì‚¬ ì–´ë¯¸ êµ¬ë‘ì  ì‚­ì œí•˜ëŠ” í•¨ìˆ˜ ë§Œë“¤ì–´ë³´ê¸°
          clean_text.append(txt)
    return " ".join(clean_text)
```

```python
# ì‹¤ì œ í…ìŠ¤íŠ¸ì— í•¨ìˆ˜ ì ìš©í•´ë³´ê¸°
okt_clean(text="ìì—°ì–´ ì²˜ë¦¬ëŠ” ì¸ê³µì§€ëŠ¥ì˜ í•œ ë¶„ì•¼ì…ë‹ˆë‹¤.")
```

## 3-1 CountVectorizer
* CountVectorizerëŠ” ì‚¬ì´í‚·ëŸ°ì—ì„œ ì œê³µí•˜ëŠ”bag of wordsë¥¼ ë§Œë“¤ ìˆ˜ ìˆëŠ” ë°©ë²•ì…ë‹ˆë‹¤.



```python
corpus = ["ì½”ë¡œë‚˜ ê±°ë¦¬ë‘ê¸°ì™€ ì½”ë¡œë‚˜ ìƒìƒì§€ì›ê¸ˆ ë¬¸ì˜ì…ë‹ˆë‹¤.",
          "ì§€í•˜ì²  ìš´í–‰ì‹œê°„ê³¼ ì§€í•˜ì²  ìš”ê¸ˆ ë¬¸ì˜ì…ë‹ˆë‹¤.",
          "ì§€í•˜ì²  ìŠ¹ê°•ì¥ ë¬¸ì˜ì…ë‹ˆë‹¤.",
          "ì½”ë¡œë‚˜ ì„ ë³„ì§„ë£Œì†Œ ë¬¸ì˜ì…ë‹ˆë‹¤.",
          "ë²„ìŠ¤ ìš´í–‰ì‹œê°„ ë¬¸ì˜ì…ë‹ˆë‹¤.",
          "ë²„ìŠ¤ í„°ë¯¸ë„ ìœ„ì¹˜ ì•ˆë‚´ì…ë‹ˆë‹¤.",
          "ì½”ë¡œë‚˜ ê±°ë¦¬ë‘ê¸° ì•ˆë‚´ì…ë‹ˆë‹¤.",
          "íƒì‹œ ìŠ¹ê°•ì¥ ë¬¸ì˜ì…ë‹ˆë‹¤."
         ]
```

```python
# CountVectorizer ì¸ìŠ¤í„´ìŠ¤ë¥¼ ë§Œë“­ë‹ˆë‹¤.
cvect = CountVectorizer()

# í…ìŠ¤íŠ¸ ë°ì´í„°ë¥¼ í•™ìŠµ(fit)í•˜ê³ , ì´ë¥¼ BOW í˜•íƒœë¡œ ë³€í™˜(transform)í•©ë‹ˆë‹¤.
X = cvect.fit_transform(corpus)

# ì´ì œ ë¬¸ì¥ë“¤ì´ ìˆ«ìë¡œ ë³€í™˜ëœ ê²°ê³¼ë¥¼ ë³´ê¸° ì¢‹ê²Œ í‘œ í˜•íƒœë¡œ ë§Œë“­ë‹ˆë‹¤.
#    - X.toarray()ëŠ” ê° ë¬¸ì¥ì„ ìˆ«ìë¡œ ë³€í™˜í•œ ê²°ê³¼ë¥¼ ë³´ê¸° ì¢‹ê²Œ ë°”ê¿”ì¤ë‹ˆë‹¤.
#    - columns=cvect.get_feature_names_out()ëŠ” ê° ì—´(ì»¬ëŸ¼)ì— í•´ë‹¹í•˜ëŠ” ë‹¨ì–´ì˜ ì´ë¦„ì„ í‘œì— ì¶”ê°€í•´ì¤ë‹ˆë‹¤.
pd.DataFrame(X.toarray(), columns=cvect.get_feature_names_out())
```
