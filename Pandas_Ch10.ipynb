# 10. apply method 활용
- 사용자가 작성한 함수를 한 번에 데이터 프레임의 각 행과 열에 적용하여 실행할 수 있게 해주는 method
- 대용량 데이터 처리시, for문보다 빠름.

## 10.1. 간단한 함수 만들기

```python
def my_sq(x):
    return x**2

def my_exp(x, n):
    return x**n

my_sq(4), my_exp(2, 4)
```

## 10.2. apply method 사용하기 기초

```python
import pandas as pd

df = pd.DataFrame({'a':[10,20,30], 'b':[20,30,40]})
df
```

```python
print(df['a']**2)
print(df['b']**2)
```

```python
sq = df['a'].apply(my_sq)
print(sq)
```

```python
ex = df['a'].apply(my_exp, n=2)
print(ex)
```

```python
ex = df['a'].apply(my_exp, n=3)
print(ex)
```

```python
df = pd.DataFrame({'a':[10,20,30], 'b':[20,30,40]})
print(df)
```

```python
def print_me(x):
    print(x)
```

```python
print(df.apply(print_me, axis=0)) # 함수를 데이터프레임의 열(0)단위로 적용할지, 행(1)단위로 적용할지 인자 설정
```

```python
# 열단위 출력 확인
print(df['a'])
print(df['b'])
```

```python
print(df.apply(print_me, axis=1)) # 행 단위
```

```python
# 행 단위 출력 확인
print(df.loc[0])
print(df.loc[1])
```

```python
def avg_3(x, y, z):
    return (x+y+z)/3
```

```python
# 오류 발생 - 3가지 인자를 전달하지 않음.
print(df.apply(avg_3))
```

```python
# 열 단위 출력
def avg_3_apply(col):
    x = col[0]
    y = col[1]
    z = col[2]
    return (x+y+z)/3

print(df.apply(avg_3_apply))
```

```python
# 인자 개수를 모를 때

def avg_3_apply(col):
    sum = 0
    for item in col:
        sum += item

    return sum/df.shape[0] #

print(df.apply(avg_3_apply))
```

```python
# 행 단위 출력
def avg_2_apply(row):
    sum = 0
    for item in row:
        sum += item
    return sum/df.shape[1]

print(df.apply(avg_2_apply, axis=1)) # 행단위로 적용하기 위해 axis 설정
```

## 10.3. apply method 사용하기 - 고급


```python
import seaborn as sns

titanic = sns.load_dataset('titanic')

print(titanic.info())
```

```python
import numpy as np

def count_missing(vec):
    null_vec = pd.isnull(vec)
    null_count = np.sum(null_vec)
    return null_count

cmis_col = titanic.apply(count_missing)

print(pd.isnull(titanic).iloc[0:5, 0:5])
print('-'*50)
print(cmis_col)
```

```python
def prop_missing(vec):
    num = count_missing(vec)
    dem = vec.size
    return num/dem

pmis_col = titanic.apply(prop_missing)
print(titanic.size)
print()
print(pmis_col)
```

```python
def prop_complete(vec):
    return 1 - prop_missing(vec)

pcom_col = titanic.apply(prop_complete)
print(pcom_col)
```

```python
cmis_row = titanic.apply(count_missing, axis=1)
pmis_row = titanic.apply(prop_missing, axis=1)
pcom_row = titanic.apply(prop_complete, axis=1)
```

```python
print(cmis_row.head())
print()
print(pmis_row.head())
print()
print(pcom_row.head())
```

```python
titanic['num_missing'] = titanic.apply(count_missing, axis=1)
print(titanic.head())
```

```python
print(titanic.loc[titanic.num_missing > 1, :].sample(5))
```
