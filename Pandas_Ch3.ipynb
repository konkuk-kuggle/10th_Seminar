2) Dataframe 만들기

```python
# DataFrame class에 dictionary 전달 - {column name : elements}
scientists = pd.DataFrame({
    'Name': ['Rosaline Franklin', 'William Gosset'],
    'Occupation': ['Chemist', 'Statistician'],
    'Born': ['1920-07-25', '1876-06-13'],
    'Died': ['1958-04-16', '1937-10-16'],
    'Age': [37, 61]})
print(scientists)
```

```python
scientists = pd.DataFrame(
    data={'Occupation': ['Chemist', 'Statistician'],
          'Born': ['1920-07-25', '1876-06-13'],
          'Died': ['1958-04-16', '1937-10-16'],
          'Age': [37, 61]},
    index=['Rosaline Franklin', 'William Gosset'],
    columns=['Occupation', 'Born', 'Age', 'Died']) #age와 died 위치 변경
print(scientists)
```

```python
#순서 보장 딕셔너리 - OrderedDict 사용
from collections import OrderedDict

scientists = pd.DataFrame(OrderedDict([
    ('Name', ['Rosaline Franklin', 'William Gosset']),
    ('Occupation', ['Chemist', 'Statistician']),
    ('Born', ['1920-07-25', '1876-06-13']),
    ('Died', ['1958-04-16', '1937-10-16']),
    ('Age', [37, 61])
]))
print(scientists)
```

## 03-5. 시리즈와 데이터프레임의 데이터 처리하기

```python
print(scientists['Born'].dtype)
print(scientists['Died'].dtype)
```

```python
# 문자열 -> datetime 자료형으로 변경
born_datetime = pd.to_datetime(scientists['Born'], format='%Y-%m-%d')
print(born_datetime)
```

```python
died_datetime = pd.to_datetime(scientists['Died'], format='%Y-%m-%d')
print(died_datetime)
```

1) 새로운 열 추가

```python
scientists['born_dt'], scientists['died_dt'] = (born_datetime, died_datetime)
```

```python
print(scientists.head())
```

```python
print(scientists.shape)
```

```python
# datetime 자료형 계산
scientists['age_days_dt'] = (scientists['died_dt'] - scientists['born_dt'])
```

2) 시리즈, 데이터프레임의 데이터 섞기

```python
import random

random.seed(42) #난수 생성 시 기준값
random.shuffle(scientists['Age']) #shuffle 메서드
print(scientists['Age'])
```

3) 데이터프레임의 열 삭제하기

```python
print(scientists.columns)
```

```python
scientists_dropped = scientists.drop(['Age'], axis=1) #axis=1 - 열 삭제4
print(scientists_dropped.columns)
```
