## 04-3. seaborn library 사용

```python
import seaborn as sns
tips = sns.load_dataset("tips")
print(tips.head())
```

##### ① 단변량 그래프 - displot & countplot

```python
sns.displot(tips["total_bill"], kde=True).fig.set_size_inches(3,2)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()

# 아래의 cell의 코드 추천
'''
`distplot` is a deprecated function and will be removed in seaborn v0.14.0.
Please adapt your code to use either `displot` (a figure-level function with
similar flexibility) or `histplot` (an axes-level function for histograms).
'''
```

```python
ax = sns.displot(tips["total_bill"], kde=True)
ax.fig.suptitle("Distribution of Total Bill")
ax.set_axis_labels(x_var="Total Bill", y_var="Frequency")
ax.fig.set_size_inches(3,2)

# sns.displot은 'FacetGrid'라는 객체를 반환 - 다중 플롯 처리
# FacetGrid 객체의 fig를 사용하여 figsize와 title를 변경해야함.
# fig.suptitle, fig.set_size_inches()...
```

```python
# rug 사용
ax = sns.displot(tips["total_bill"], kde=True, rug=True)
ax.fig.suptitle("Distribution of Total Bill")
ax.set_axis_labels(x_var="Total Bill", y_var="Frequency")
ax.fig.set_size_inches(3,2)
```

```python
# sns.countplot은 Axes라는 객체로 반환 - plt.subplot와 동일(Fig객체, Axes객체)
ax = plt.subplots(figsize=(3,3))
ax = sns.countplot(x='day', data=tips)
ax.set_title("Count of Days")
ax.set_xlabel("Day of the Week")
ax.set_ylabel("Frequency")
ax
```

##### ② 이변량 그래프 - regplot, jointplot, barplot, boxplot, violintplot

```python
ax = plt.subplots(figsize=(3,3))
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=True) #산점도 + 회귀선
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
ax
```

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.fig.suptitle('Joint Plot of Total Bill and Tip', y=1.1) #JointGrid 객체 반환
joint.fig.set_size_inches(3,3)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip') # displot는 x_var, y_var
joint
```

```python
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', y=1.1, fontsize=10)
hexbin.fig.set_size_inches(3,3)
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
```

```python
ax = plt.subplots(figsize=(3,3))
ax = sns.kdeplot(data=tips,
                 x=tips['total_bill'], #data, data2 사용X
                 y=tips['tip'],
                 fill=True) # shade 사용X
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
ax
```

```python
ax = plt.subplots(figsize=(3,3))
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Barplot of Average Total Bill for tiem of day')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Average Total Bill')
ax
```

```python
ax = plt.subplots(figsize=(3,3))
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of Total Bill by Time of Day')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Total Bill')
ax
```

```python
ax = plt.subplots(figsize=(3,3))
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Violin Plot of Total Bill by Time of Day')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Total Bill')
ax
```

```python
fig = sns.pairplot(tips)
fig.fig.suptitle('Pairplot of Tips Dataset', y=1.02, fontsize=10)
fig.fig.set_size_inches(5,5)
fig  # 그래프 중복 - PairGird 사용
```

```python
# 책에서는 변수를 계속 선언했으나, 실제로는 X
pair_grid = sns.PairGrid(tips)
pair_grid.map_upper(sns.regplot)
pair_grid.map_lower(sns.kdeplot)
pair_grid.map_diag(sns.histplot, kde=True)
pair_grid.fig.suptitle('Pairplot of Tips Dataset', y=1.02, fontsize=10)
pair_grid.fig.set_size_inches(5,5)
```

## 04-5. seaborn library로 그래프 스타일 설정하기

```python
fig, ax = plt.subplots(figsize=(3,3))
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(3,3))
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
fig = plt.figure(figsize=(5,5))

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)
fig.tight_layout()
```
