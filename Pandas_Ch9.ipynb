##### splitlines method

```python
# 여러 행을 가진 문자열을 분리하고 리스트로 반환
# """___""" : 큰 따옴표를 3쌍 사용하면 여러 행의 문자열을 선언할 수 있음.

multi_str = """Guard: What? Ridden on a horse?
King Arthur: Yes!
Guard: You're using coconuts!
King Arthur: What?
Guard: You've got ... coconut[s] and you're bangin' 'em together.
"""

print(multi_str)
print()
print(multi_str.splitlines())
```

## 9.3. 문자열 포맷팅

### 9.3.1. 문자열 포맷팅하기

출력할 문자열의 형식을 지정하거나 변수를 조합하여 출력

```python
# {} : place holder. format method와 함께 사용하여 특정 문자열 형식을 지정하고 원하는 변수를 넣을 수 있도록 함.

var = 'flesh wound'
s = "It's just a {}!"

print(s.format(var))
print()
print(s.format('scratch'))
```

```python
# 여러 개의 place holder
s = """Black Knight: 'Tis but a {0}.
King Arthur: A {0}? Your arm's off!
"""

print(s.format('scratch'))
```

```python
# {}에 특정 변수를 지정하여 사용
s = 'Hayden Planetarium Coordinates: {lat}, {lon}'
print(s.format(lat='40.7815° N', lon='73.9733° W'))
```

```python
## 참고 - f-string

year = 2024
name = 'chaehyun'

print(f'{year} Jung {name}')

print('{y} Jung {n}'.format(y=year, n=name)) # 순서가 맞으면 변수 지정 필요 없음.
```

### 9.3.2. 숫자 데이터 포맷팅하기

```python
print('Some digits of pi: {}'.format(3.14159265359))
```

```python
# 쉼표 지정
print("In 2005, Lu Chao of China recited {:,} digits of pi".format(67890))
```

```python
# 유효숫자 지정 + 백분율 환산
# n: .m - n은 format의 n번째 숫자, m은 유효숫자
print("I remember {0:.4} or {0:.4%} of what Lu Chao recited".format(7/67890))
print("{1:.4} and {0:.3}".format(7/67890, 1/67890))
```

```python
# 자리수 지정 - d:decimal
print("My ID number is {0:05d}".format(42))
```

### 9.3.3. %연산자로 포맷팅하기

```python
# %d : 10진수
s = 'I only know %d digits of pi' % 7
print(s)
```

```python
# %s : string
# %.nf : 소수점 아래 n자리까지 표현한 float

print('Some digits of %(cont)s: %(value).2f' % {'cont': 'e', 'value': 2.718})
```

## 9.4. 정규식으로 문자열 처리에 날개 달기

```python
import re

# Sample test strings
test_string1 = " abc on like 1234567 apple application"
test_string2 = "The number is 010-1234-5678"
test_string3 = "cfhball"
```

#### ① regex patterns

. ^ $ * + ? {m} {m,n} \ [] | ()


```python
# .a : 'a' 앞에 문자가 1개 있는 패턴
# findall : 모든 패턴을 찾아 리스트로 반환
print(re.findall(r'.a', test_string1))  # '_abc','_apple', '_application', 'applic_ation'

# ^ a : 문자열의 처음부터 ' a'와 일치하는 패턴 찾기
print(re.findall(r'^ a', test_string1))

# on$ :  문자열의 끝 부분에 'on'과 일치하는 패턴 찾기
print(re.findall(r'on$', test_string1))
```

```python
# abc? : ab이거나 abc인 패턴 = ?앞의 문자가 있거나 없음.
print(re.findall(r'apple?', test_string1))  # Output: ['apple', 'appl']
```

```python
# n\d* : 'n' 다음에 숫자가 0개 이상인 패턴
print(re.findall(r' \d*', test_string2)) # _number, _is, _010

# \d+ : 숫자가 한 개 이상인 패턴
print(re.findall(r'\d+', test_string2))  # Output: ['010', '1234', '5678']

# n\d+ : 'n' 다음에 숫자가 한 개 이상인 패턴
print(re.findall(r'-\d+', test_string2))   # Output: ['-1234', '-5678']
```

```python
# \d{m} : 숫자가 m개 있는 패턴
print(re.findall(r'\d{2}', test_string2))  # Output: ['01', '01', '23', '45', '67', '78']

# \d{m,n} - 숫자가 m개 이상, n개 이하 있는 패턴
print(re.findall(r'\d{2,4}', test_string2))  # Output: ['010', '1234', '5678']
```

```python
# \특수문자 - escape 문자 뒤의 '특수문자'를 검색
print(re.findall(r'\*', "1 * 2 = 2"))  # Output: ['*']
```

```python
# [abc] : a, b, c 중 한 개를 포함하는 패턴
print(re.findall(r'[cfh]', test_string3))  # Output: ['c', 'f', 'h']

# apple | application : apple이거나 application 중 하나만 일치하는 패턴 (or)
print(re.findall(r'apple|application', test_string1))  # Output: ['apple', 'application']

# (\d+)-(\d+)-(\d+) : 해당 패턴을 만족하는 괄호 안의 패턴을 튜플로 반환
print(re.findall(r'(\d+)-(\d+)-(\d+)', test_string2))  # Output: [('010', '1234', '5678')]
```

#### ② Special Characters

```python
# \d - 모든 숫자 문자 1개 = [0-9]
print(re.findall(r'\d', test_string2))  # 1012345678

# \D - 숫자 이외의 문자 1개 = [^0-9]
print(re.findall(r'\D', test_string2))  # "The number is - -"

# \s - 모든 공백 문자 (space, tab, etc.)
print(re.findall(r'\s', test_string2))  # "The_number_is_010-1234-5678"

# \S - 공백 문자 이외의 문자 1개
print(re.findall(r'\S', test_string2[-17:])) # ' is 010-1234-5678'

# \w - 알파벳 문자 1개
print(re.findall(r'\w', test_string2[:10]))

# \W - 알파벳 이외의 문자 1개
print(re.findall(r'\W', test_string2))
```

#### ③ regex methods

```python
# search() - Find the first match
m = re.search(r'[0-9]{4}', test_string2) # [0에서 9 사이]의 숫자가 {4}번 반복하는 첫번째 패턴의 양 끝 인덱스 반환
print(m, m.group())

# match() - Check for a match at the start of the string
m = re.match(r'[0-9]{3}', test_string2)
print(m)  # 문자열이 해당 패턴으로 시작하지 않음.

# fullmatch() - Check if the entire string matches the pattern
m = re.fullmatch(r'\d+\D\d+\D\d+', '010-1234-5678')
print(m, bool(m))

# split() - Split string by the given pattern
print(re.split(r'\s', test_string2))  # 공백문자를 기준으로 잘라냄.

# findall() - Find all occurrences of the pattern
print(re.findall(r'[0-9]{4}', test_string2))  # 숫자가 4번 반복하는 패턴

# finditer() - Return an iterator yielding match objects
for m in re.finditer(r'[0-9]{4}', test_string2):
    print(m.group())  # '1234', '5678'

# sub() - Replace occurrences of the pattern
print(re.sub(r'\D\d{4}\D', '-easypsub(전달한 값)-', test_string2))  # Output: '010-easypsub(전달한 값)-5678'
```

#### ④ Examples

```python
import re

tele_num = '1234567890'
```

```python
m = re.match(pattern='\d\d\d\d\d\d\d\d\d\d', string=tele_num)
print(type(m))
print(m)
```

```python
print(bool(m))
```

```python
if m:
    print('match')
else:
    print('no match')
```

```python
print(m.start()) # 찾은 패턴 첫번째 인덱스
print(m.end())   # 찾은 패턴 마지막 인덱스
print(m.span())  # 찾은 패턴 첫번째 인덱스와 마지막 인덱스 한 번에 반환
print(m.group()) # 찾아낸 패턴 반환
```

```python
tele_num_spaces = '123 456 7890'
m = re.match(pattern='\d{10}', string=tele_num_spaces)
print(m)
```

```python
if m:
    print('match')
else:
    print('no match')
```

```python
p = '\d{3}\s?\d{3}\s?\d{4}'
m = re.match(pattern=p, string=tele_num_spaces)
print(m)
```

```python
# 지역 코드 존재
tele_num_space_paren_dash = '(123) 456-7890'
p = '\(?\d{3}\)?\s?\d{3}\s?-?\d{4}'
m = re.match(pattern=p, string=tele_num_space_paren_dash)
print(m)
```

```python
# 국가 코드 존재
cnty_tele_num_space_paren_dash = '+1 (123) 456-7890'
p = '\+?1\s?\(?\d{3}\)?\s?\d{3}\s?-?\d{4}'
m = re.match(pattern=p, string=cnty_tele_num_space_paren_dash)
print(m)
```

```python
# compile method - 패턴을 컴파일하여 변수에 저장하고 반복해서 사용
p = re.compile('\d{10}')
s = '1234567890'
m = p.match(s)      # re.match(pattern, string)
print(m)
```
