```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/yeong-eun/Desktop/영은/쿠글/study_performance.csv')
df = df[['race_ethnicity', 'parental_level_of_education', 'math_score', 'reading_score', 'writing_score']]

list = df['race_ethnicity'].drop_duplicates()

groupA = df[df['race_ethnicity'] == 'group A'].reset_index(drop=True)
groupB = df[df['race_ethnicity'] == 'group B'].reset_index(drop=True)
groupC = df[df['race_ethnicity'] == 'group C'].reset_index(drop=True)
groupD = df[df['race_ethnicity'] == 'group D'].reset_index(drop=True)
groupE = df[df['race_ethnicity'] == 'group E'].reset_index(drop=True)

plt.style.use('default')
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

def plot(ax, df, title):

    xticks = ['수학', '읽기', '쓰기']
    ax.boxplot([df['math_score'], df['reading_score'], df['writing_score']])

    ax.set_xticks([1, 2, 3])
    ax.set_xticklabels(xticks, rotation=0, ha='center')

    ax.set_yticks(np.arange(0, 101, 10))
    ax.set_ylim(0, 101)
    ax.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, zorder=1)

    ax.set_xlabel('과목', fontweight='bold')
    ax.set_ylabel('점수', fontweight='bold')
    ax.set_title(title, fontweight='bold', fontsize=15)

fig, axes = plt.subplots(1, 5, figsize=(26, 7), dpi=300)

plot(axes[0], groupA, 'Group A')
plot(axes[1], groupB, 'Group B')
plot(axes[2], groupC, 'Group C')
plot(axes[3], groupD, 'Group D')
plot(axes[4], groupE, 'Group E')

plt.tight_layout()
plt.show()
```
