### 7.2.2. 하나의 열 분리하기

```python
# split method : Splits the string in the Series/Index from the beginning, at the specified delimiter(regular expression) string.
variable_split = ebola_long.variable.str.split('_')
```

- __*accessor*__   
: 데이터 타입에 대해 고유한 작업이나 메서드를 제공하는 인터페이스  
ex) .str   
: attribute와 다름.   
:  속성처럼 '*.*' 연산자를 통해 접근할 수 있지만, 속성보다는 더 특수화된 작업을 위한 도구   
   
- __*attribute*__   
: Python 객체(object)에 저장된 값이나 속성, 상태를 나타내는 변수   
: 객체의 . 연산자를 통해 접근

```python
print(type(variable_split))       # Series로 반환
print(type(variable_split[0]))    # 각 요소는 list로 반환
```

```python
# get 함수 : Extract element from lists, tuples, dict, or strings in each element in the Series/Index.
status_values = variable_split.str.get(0)   # 상태 부분 추출
country_values = variable_split.str.get(1)  # 나라 부분 추출
```

## 7.3. 여러 열을 하나로 정리하기

```python
weather = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kuggle/data/data/weather.csv')
print(weather.iloc[:5, :])
```

```python
# 일별 온도 측정값(d1, d2, ..., d31)을 피벗
weather_melt = pd.melt(weather, id_vars=['id', 'year', 'month', 'element'], var_name='day', value_name='temp')
print(weather_melt.head())
```

```python
# pivot_table method : Create a spreadsheet-style pivot table as a DataFrame.
# (index=None, columns=None, aggfunc: AggFuncType='mean', dropna: bool=True,, observed: bool=False, sort: bool=True)
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'], #고정할 열
    columns='element',                    #피벗할 열
    values='temp',
    dropna=False
)
print(weather_tidy)
print(weather_tidy.columns)
```

```python
# reset_index method : index를 새로 지정
weather_tidy_flat = weather_tidy.reset_index()
print(weather_tidy_flat.head())
print(weather_tidy_flat.shape)
# 원래 index를 풂. - column 수 증가
```

## 7.4. 중복 데이터 처리하기

```python
billboard = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Kuggle/data/data/billboard.csv')
billboard_long = pd.melt(billboard, id_vars=['year', 'artist', 'track', 'time', 'date.entered'], var_name='week', value_name='rating')

print(billboard_long.shape)
print()
print(billboard_long.head())
```

```python
# 중복데이터 확인 - year, artist, track, time, date 중복
print(billboard_long[billboard_long.track == 'Loser'].head())
```

```python
billboard_songs = billboard_long[['year', 'artist', 'track', 'time']]
print(billboard_songs.shape)
```

```python
# drop_duplicates method : Return DataFrame with duplicate rows removed.
billboard_songs = billboard_songs.drop_duplicates()
print(billboard_songs.shape)
```

```python
# 노래 별로 index 지정
billboard_songs['id'] = range(len(billboard_songs)) # 순차적으로 index 지정
print(billboard_songs.head(n=10))
```

```python
# merge method
# '주간 순위 데이터'와 '노래 정보 데이터' 합침. - 주간 순위 데이터에 노래별로 id 생성
billboard_ratings = billboard_long.merge(billboard_songs, on=['year', 'artist', 'track', 'time'])
print(billboard_ratings.shape)
print()
print(billboard_ratings.head())
```

os.path.join을 사용할 때, 첫 번째 인자로 빈 문자열 ''을 전달하면 다음과 같은 상황에서 유용


1. 절대 경로로 시작할 때   
: os.path.join은 첫 번째 인자가 절대 경로라면, 이후의 모든 경로 조각들을 무시하고, 절대 경로로 시작하는 경로를 반환함. 이때 빈 문자열이 사용되면, 결합된 결과가 올바르게 작동하도록 도와줌.

2. 유연한 경로 결합   
: 빈 문자열을 첫 번째 인자로 넣으면, 경로 조합을 할 때 코드의 유연성을 높일 수 있음. 빈 문자열은 경로 결합에 영향을 미치지 않기 때문에, 조건에 따라 경로의 일부가 추가되거나 제거될 때 유용함.
